```python
import requests
import json
```

```python
# Zip code less Maryland and Delaware 
zip = ('34953', '31909', '42701', '71301', '39503', '28806', '74145', '29926','38506', '79936', '24605', '26501')

for i in zip:
    api_url = f'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{i}/2014-01-01/2014-12-31?unitGroup=metric&include=days&key=WU4N9BE773YG7WHMSAZNUHCM2&contentType=json'

    response = requests.get(api_url)
    print(response.json())


    json_string = json.dumps(response.json())
    with open('C:/Users/kgame/Downloads/Class.Economic Analytics/weather ' + i + '.json', 'w')  as outfile:
        outfile.write(json_string)
```

```python
import glob

read_files = glob.glob("C:/Users/kgame/Downloads/Class.Economic Analytics/weather*")
output_list = []
zip = ['36830','34953', '31909', '42701', '71301', '39503', '28806', '74145', '29926','38506', '79936', '24605', '26501']
for f in read_files:
    with open(f, "rb") as infile:
        output_list.append(json.load(infile))


all_items = []
i = 0
n = 0
while (i < len(zip)):
    for json_file in output_list:
        for n in range(365):
            json_file['days'][n]['zip'] = zip[i] 
            del json_file['days'][n]['stations'] 
        all_items += json_file['days']   
        i += 1

textfile_merged = open('C:/Users/kgame/Downloads/Class.Economic Analytics/merged_weather.json', 'w')
json.dump( all_items , textfile_merged)
textfile_merged.close()
```
